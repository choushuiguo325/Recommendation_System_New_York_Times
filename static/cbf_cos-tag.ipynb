```python
import pandas as pd
import numpy as np
```

```python
attributes_df = pd.read_csv('attributes.csv')
total_articles = list(attributes_df['contentID'])
pvs = pd.read_csv('pvs.csv')
visits = pd.read_csv('visits.csv')
```

```python
# inner join visits and page view datasets
# user visits with time information
user_visits = pd.merge(visits,pvs, on=('visitorID','visitID'),how='inner')[['visitID','visitorID','contentID','visitDateTime']]
user_visits = user_visits[user_visits['contentID'].isin(total_articles)]
user_visits.head()
```

```python
# split the datasets into train and test datasets 
train_utility = user_visits.loc[user_visits['visitDateTime'] < '2018-09-16'].drop(columns='visitDateTime')
# user profile for train set
train_utility = train_utility.groupby(['visitorID','contentID']).count()
train_utility = train_utility.reset_index()
train_utility = train_utility.rename(columns={'visitID':'visitCount'})
train_utility
```

```python
# split the datasets into train and test datasets 
test_utility = user_visits.loc[user_visits['visitDateTime'] >= '2018-09-16'].drop(columns='visitDateTime')
# user profile for test set
test_utility = test_utility.groupby(['visitorID','contentID']).count()
test_utility = test_utility.reset_index()
test_utility = test_utility.rename(columns={'visitID':'visitCount'})
test_utility
```

```python
popular_articles = test_utility[['visitorID','contentID']].groupby('contentID').count().sort_values('visitorID',ascending=False).head(5)
popular_articles = list(popular_articles.index)
```

```python
# create users and articles lists for train ans test sets
train_articles = list(train_utility['contentID'].unique())
train_users = list(train_utility['visitorID'].unique())
print('train articles:',len(train_articles))
print('train users:',len(train_users))
test_articles = list(test_utility['contentID'].unique())
test_users = list(test_utility['visitorID'].unique())
print('test articles:',len(test_articles))
print('test users:',len(test_users))
```

```python
# create similarity matrix between all the articles and test articles
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine
cleared_attributes_df = attributes_df.set_index(attributes_df.iloc[:,0]).iloc[:,1:]
similarity_df = pd.DataFrame(1-pairwise_distances(np.array(cleared_attributes_df), metric="cosine"))
similarity_df.index = total_articles
similarity_df.columns = total_articles
# ---------------
similarity_df['counts'] = attributes_count
similarity_df = similarity_df.iloc[similarity_df.index.isin(test_articles)]
similarity_df
```

```python
similarity_df.sort_values('www.arkansasonline.com/news/2018/aug/23/sheriffs-office-dad-accidentally-kills-14-year-old/', ascending = False)
```

```python
attributes_articles_index = list(similarity_df.index)
```

```python
# top-5 articles
n=0
blank_df = pd.DataFrame({'user':[], 'read':[],'unread': [], 'similarity':[], 'counts':[]})
total_articles = blank_df
for user in test_users:
#     if the user is not a new user
    if user in train_users :
        # filtered out the artilces users have already read
        read_articles = list(train_utility.loc[train_utility['visitorID']==user]['contentID'])
        n += 1
        print(n)
        # calculate the similarity between unread articles and read articles
        new_articles = blank_df
        for article in read_articles:
            one_article = pd.DataFrame({'user':user, 'read': article,'unread': attributes_articles_index, 'similarity':similarity_df[article],'counts':similarity_df['counts']})
            one_article.index = attributes_articles_index
            one_article = one_article[one_article.unread.isin(read_articles) == False]
            one_article = one_article.sort_values(['similarity','counts'],ascending=False)
            one_article = one_article.head(5)
            new_articles = new_articles.append(one_article) 
        # top-5 similarity results for one user
        new_articles = new_articles.sort_values('similarity',ascending=False).head(5)

    else: 
        new_articles = pd.DataFrame({'user':user, 'read':[0,0,0,0,0],'unread':popular_articles, 'similarity':[0,0,0,0,0], 'counts':[0,0,0,0,0]})
    # combine the similarity results for all users
    total_articles = total_articles.append(new_articles)
    
total_articles
```

```python
total_articles
```

```python
top5rec = total_articles[['user','unread']].rename(columns={'user':'visitorID','unread':'contentID'}).reset_index(drop=True)
top5rec.head(20)
```

```python
actual_articles = test_utility[['visitorID','contentID']]
actual_articles['actual'] = actual_articles['contentID']
top5rec['predicted'] = top5rec['contentID']
results5 = pd.merge(top5rec,actual_articles, on=['visitorID','contentID'],how ="left")
results5['match'] = np.where(results5['predicted'] == results5['actual'], 0.2, 0)
results5.head(20)
```

```python
scores = results5[['visitorID','match']].groupby(['visitorID']).sum()
scores
```

```python
precision = scores['match'].mean()
precision
```
