```python
import pandas as pd
import numpy as np
```

```python
attributes_df = pd.read_csv('cal_attributes.csv')
attributes_df = attributes_df.set_index(attributes_df.iloc[:,0]).iloc[:,1:]
attributes_df
```

```python
utility_df = pd.read_csv('cal_utility.csv')
utility_df = utility_df.set_index(utility_df.iloc[:,0]).iloc[:,1:]
utility_df = utility_df.fillna(0)
utility_df
```

```python
np.transpose(utility_df)
```

```python
user_profile = np.matmul(np.array(np.transpose(utility_df)), np.array(attributes_df))
user_profile
```

```python
user_profile.shape
```

```python
attributes_df.shape
```

```python
predicted_score = np.matmul(np.array(user_profile), np.array(np.transpose(attributes_df)))
predicted_score
```

```python
document_lists = list(utility_df.index)
user_lists = list(utility_df.columns)
```

```python
results = pd.DataFrame({'document':[], 'user':[], 'scores':[]})
for i in range(0,len(user_lists)):
    new_df = pd.DataFrame({'document':document_lists, 'user':user_lists[i], 'scores':predicted_score[i]})
    new_df = new_df.set_index(new_df['document'])
    results = results.append(new_df)
```

```python
sorted_results = results.groupby(['user'])['scores'].nlargest(3)
sorted_results
```
