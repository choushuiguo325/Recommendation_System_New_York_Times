```python
import pandas as pd
import numpy as np
```

```python
attributes_df = pd.read_csv('cal_attributes.csv')
attributes_df = attributes_df.set_index(attributes_df.iloc[:,0]).iloc[:,1:]
attributes_df
```

```python
utility_df = pd.read_csv('cal_utility.csv')
utility_df = utility_df.set_index(utility_df.iloc[:,0]).iloc[:,1:]
utility_df = utility_df.replace(-1,1).fillna(0)
utility_df
```

```python
document_lists = list(utility_df.index)
user_lists = list(utility_df.columns)
```

```python
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine
```

```python
similarity_df = pd.DataFrame(1-pairwise_distances(np.array(attributes_df), metric="cosine"))
similarity_df.index = document_lists
similarity_df.columns = document_lists
similarity_df
```

```python
total_articles = pd.DataFrame({'user':[], 'read':[],'unread': [], 'similarity':[]})
for user in user_lists:
    # filtered out the artilces users have already read
    filtered = utility_df.loc[utility_df[user] ==1][user]
    user_articles = list(filtered.index)
    # calculate the similarity between unread articles and read articles
    new_articles = pd.DataFrame({'user':[], 'read':[],'unread': [], 'similarity':[]})
    for article in user_articles:
        one_article = pd.DataFrame({'user':user, 'read':article,'unread': document_lists, 'similarity':similarity_df[article]})
        one_article.index = document_lists
        new_articles = new_articles.append(one_article)
        new_articles = new_articles.drop(index=article)
    # top-3 similarity results for one user
    new_articles = new_articles.sort_values('similarity',ascending=False).head(3)
    print(new_articles)
    # combine the similarity results for all users
    print()
    total_articles = total_articles.append(new_articles)
    
total_articles
```
